```python
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np
import pytesseract
```

```python
captcha_list = os.listdir('captcha_data')
captcha_list[3]
```

```python
img = cv2.imread(f'captcha_data/{captcha_list[4]}')
plt.imshow(img)
```

```python
kernel = np.ones((4, 4), np.uint8)
a = cv2.erode(img, kernel, iterations=1)
plt.imshow(a)
```

```python
b = cv2.GaussianBlur(a, (5, 5), 0)
plt.imshow(b)
```

```python
c = cv2.Canny(b, 30, 150)
plt.imshow(c)
```

```python
d = cv2.dilate(c, kernel, iterations=1)
plt.imshow(d)
```

```python
e = cv2.morphologyEx(d,cv2.MORPH_GRADIENT,np.ones((5, 5), np.uint8))
plt.imshow(e)
```
