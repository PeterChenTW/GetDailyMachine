```python
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
captcha_list = os.listdir('captcha_data')
captcha_list[0]
```

```python
img = cv2.imread(f'captcha_data/{captcha_list[0]}')
plt.imshow(img)
```

```python
kernel = np.ones((4, 4), np.uint8)
img = cv2.erode(img, kernel, iterations=1)
plt.imshow(img)
```

```python
img = cv2.GaussianBlur(img, (5, 5), 0)
plt.imshow(img)
```

```python
img = cv2.Canny(img, 30, 150)
plt.imshow(img)
```

```python
img = cv2.dilate(img, kernel, iterations=1)
plt.imshow(img)
```
